```python
class banking_account:
    def __init__(self):
        self.a=int(input("Enter A/c No.:"))
        self.b=input("Enter Your Name: ")
        self.c=input("Enter Type of A/c You Want: ")
        self.d=int(input("Enter Amount to Deposit: "))
        print("Your A/c has been created successfully")
     
    #Balance Checking
    def bal(self,acno):
        k=0
        if(self.a==acno):
            print("A/c No.:",self.a)
            print("Name:",self.b)
            print("Type of A/c:",self.c)
            print("Amount:",self.d)
            print("------------------")
            k=1
        return k
    
    #Deposit 
    def depo(self,acno):
        k=0
        if(self.a==acno):
            A=int(input("The amount to be deposited:"))
            x=self.d+A
            print("A/c No.:",self.a)
            print("Name:",self.b)
            print("Type of A/c:",self.c)
            print("Amount:",x)
            print("------------------")
            k=1
        return k
    
    #Withdraw 
    def witd(self,acno):
        k=0
        if(self.a==acno):
            A=int(input("The amount to be withdraw:"))
            x=self.d-A
            print("A/c No.:",self.a)
            print("Name:",self.b)
            print("Type of A/c:",self.c)
            print("Amount:",x)
            print("------------------")
            k=1
        return k
    
ls=[]
n=int(input("How many A/c you want to create?")) #No. of A/c to be created
for i in range(0,n):
    b=banking_account()
    ls.append(b)
    
        
print("1.Deposit\n2.Withdraw\n3.Balance Enquiry.") #Operation after creating A/c
    
ch=int(input("Enter your choice:"))
    
if(ch==3):
        k=0
        acno=int(input("Enter A/c No. to check balance:"))
        for i in range(0,n):
            if(ls[i].bal(acno)==1):
                k=k+1
            if(k==0):
                print("Not found")
                
elif(ch==1):
        k=0
        acno=int(input("Enter A/c No. to check balance:"))
        for i in range(0,n):
            if(ls[i].bal(acno)==1):
                k=k+1
            if(k==0):
                print("Not found")
    
else:
        k=0
        acno=int(input("Enter A/c No. to check balance:"))
        for i in range(0,n):
            if(ls[i].bal(acno)==1):
                k=k+1
            if(k==0):
                print("Not found")
```
